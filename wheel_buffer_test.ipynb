```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Координаты вершин квадрата (по часовой стрелке)
points = [
    (0, 0),    # левая нижняя
    (1, 0),    # правая нижняя
    (1, 1),    # правая верхняя
    (0, 1)     # левая верхняя
]

# Нарисуем только 3 стороны (без верхней)
# Нижняя грань
ax.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], color='black')
# Правая грань
ax.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], color='black')
# Левая грань
ax.plot([points[3][0], points[0][0]], [points[3][1], points[0][1]], color='black')

ax.set_aspect('equal')
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.2, 1.2)
plt.show()
```
